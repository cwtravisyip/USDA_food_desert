```python
import pandas as pd 
import numpy as np
```

```python
df = pd.read_csv("census_2010.csv")
df.drop(columns =["Unnamed: 0"], inplace = True)
```

```python
df.set_index('CensusTract', inplace = True)
df.head()
```

```python
df.describe()
```

```python
# replacing encoded values as nan and dropping the row
df.replace([-666666666,-999999999,-88888888], np.nan, inplace = True)
df.dropna(inplace  = True)
```

```python
df.columns
```

### Computing the Percentage

```python
# compute the share of population by ethnic group
for x in ['P001002', 'P001003', 'P001004', 'P001005', 'P001006', 'P001007','P001008', 'P001009']:
    df[x+"PE"] = df[x]/df['P001001']*100
    df.drop(columns =x, inplace = True)
```

```python
#compute Unemloyment Rate
df['DP03_UNEMPLOY'] = df['DP03_0005PE'] / df['DP03_0002PE'] *100
```

```python
# compute percentage tenure type
for x in ['H004002', 'H004003', 'H004004']:
    df[x+"PE"] = df[x]/df['H004001'] *100
    df.drop(columns =x, inplace = True)
```

```python
df.columns
```

### Dropping columns to avoid multicollinearity

```python
cols = ['DP03_0002PE',
        'DP03_0003PE',
        'DP03_0004PE',
        'DP03_0005PE',
        'H003001',
        'H003002',
        'H003003',
        'H004001',
        'H013001',
        'H013002',
        'H013003',
        'H013004',
        'H013005',
        'H013006',
        'H013007',
        'H013008',
        'P001001',
        'P001002PE']
df.drop(columns = cols, inplace = True)
```

```python
df.columns
```

### Check rows with 1 or more NaN value

```python
# inspecting rows with null data
df[df.isna().sum(axis = 1) > 0]
```

```python
# dropping the rows with null value
drop = df[df.isna().sum(axis = 1) > 0].index
df.drop(index = drop,inplace = True)
```
